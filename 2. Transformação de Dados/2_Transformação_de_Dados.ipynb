# 1. Importar Bibliotecas

```python
# Importar a biblioteca Pandas
import pandas as pd
```

# 2. Carregar os arquivos para o Colab

```python
# Carregar o arquivo sales_data.csv dentro de uma variável "df"(dataframe) e verificar as colunas e primeiras linhas do dataframe

df_sales = pd.read_csv('sales_data.csv', sep=";")
df_customers = pd.read_csv('customer_data.csv', sep=";")
```

# 3. Mostrar as primeiras linhas de cada tabela com suas colunas

```python
# O camando .head vai mostrar as primeiras linhas de cada dataframe, adicionei o "print" apenas para organizar a visualização
print("Vendas")
df_sales.head()
```

```python
# O camando .head vai mostrar as primeiras linhas de cada dataframe, adicionei o "print" apenas para organizar a visualização
print("\nClientes")
df_customers.head()
```

# 4. Renomear a coluna "id_cliente" da tabela de Clientes

```python
# Como a coluna na tabela de vendas é "cliente_id" e na tabela de clientes é "id_cliente", é bom alterar para evitar quaisquer erros. Fiz isso usando a função .rename(columns)
df_customers.rename(columns={'id_cliente': 'cliente_id'}, inplace=True)
```

```python
# Verificar os nomes das colunas após a renomeação
print("\nDados de Clientes (Após Renomeação):")
print(df_customers.head())
```

# 5. Combinar os dataframes

```python
# Agora que o nome da coluna foi alterado, posso combinar os dataframes usando a chave comum "cliente_id"
df_combined = pd.merge(df_sales, df_customers, on='cliente_id')
```

```python
# Vou usar a função .head para mostrar as primeiras linhas desse novo dataframe combinado
print("\nDados Combinados:")
print(df_combined.head())
```

# 6. Criar um novo .csv com os dados combinados

```python
# Achei interessante criar um arquivo .csv e outro em excel com os dados combinados, como um double check de que foi feito corretamente
output_file_path = '/content/combined_data.csv'
output_file_path_excel = '/content/combined_data.xlsx'
df_combined.to_csv(output_file_path, index=False)
df_combined.to_excel(output_file_path_excel, index=False)
```

# 7. Ver informações sobre o dataframe combinado para iniciar uma análise exploratória

```python
print("\nInformações do DataFrame Combinado:")
df_combined.info()

print("\nEstatísticas Descritivas do DataFrame Combinado:")
df_combined.describe()
```
